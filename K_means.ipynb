### Step 23: Encoding Categorical Data
Use **LabelEncoder** to convert the "Genre" column (which is categorical) into numerical values. This step is necessary for preparing the dataset for machine learning models, which require numerical input. The `fit_transform()` method is applied to encode the values in the "Genre" column.


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])
df.head()
```

### Step 24: Defining the Feature Set
Define the feature set `X` by selecting all columns starting from the 4th column onwards (excluding the first three columns: "CustomerID," "Genre," and "Age"). This step prepares the data for clustering by isolating the input features for the K-means model.


```python
X=df.iloc[:,3:]
X.head()
```

### Step 25: Applying K-Means Clustering
Apply the **K-Means** algorithm to the feature set `X`, with 5 clusters (`n_clusters=5`). The model is trained with 20 initializations (`n_init=20`) to ensure stable convergence. The resulting cluster labels are added to the dataset as a new column, `k_cluster`, which indicates the cluster each customer belongs to.


```python
from sklearn.cluster import KMeans
cluster=KMeans(n_clusters=5,random_state=0,n_init=20)
df['k_cluster']=cluster.fit_predict(X)
df.head()
```

### Step 26: Evaluating the Clustering Performance with Silhouette Score
Calculate the **Silhouette Score** to evaluate the quality of the clustering. The silhouette score ranges from -1 to 1, where a higher score indicates better-defined clusters. This step helps assess how well the K-means algorithm has grouped the customers into clusters.


```python
from sklearn.metrics import silhouette_score
score=silhouette_score(X,df['k_cluster'])
score
```

### Step 27: Determining the Optimal Number of Clusters
Evaluate the performance of the K-Means clustering for different numbers of clusters (from 2 to 10) by calculating the **Silhouette Score** for each case. This step helps identify the optimal number of clusters by selecting the value that yields the highest silhouette score, ensuring the best possible clustering.


```python
scores=[]
for i in range(2,11):
  cluster=KMeans(n_clusters=i,random_state=0,n_init=20)
  df['k_cluster']=cluster.fit_predict(X)
  scores.append(silhouette_score(X,df['k_cluster']))
print(scores)
```

### Step 28: Visualizing Silhouette Scores for Different Cluster Counts
Plot a line graph and scatter plot to visualize the silhouette scores for different numbers of clusters (from 2 to 10). This helps in identifying the optimal number of clusters visually, where the highest score corresponds to the best clustering configuration.


```python
plt.scatter(range(2,11),scores)
plt.plot(range(2,11),scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
