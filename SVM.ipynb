### Step 10: Removing Duplicate Entries

The `drop_duplicates()` method is used to remove any duplicate rows from the dataset. This ensures that the model is trained on unique data, preventing biases or inaccuracies caused by redundant entries.


```python
df.drop_duplicates(inplace=True)
```

### Step 11: Separating Features and Target

The `label` column is dropped from the dataset to create a new DataFrame (`df_cont`) that contains only the feature variables. This step is essential to separate the independent variables (features) from the dependent variable (target), which is used for training the model.


```python
df_cont=df.drop(df[['label']],axis=1)
```

### Step 12: Standardizing the Features

The `StandardScaler` is used to standardize the features by scaling them to have a mean of 0 and a standard deviation of 1. This step ensures that all features contribute equally to the model, especially important for algorithms like SVM. The scaled features are then converted into a DataFrame for further analysis.


```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(df_cont)
SS_X=pd.DataFrame(SS_X,columns=df_cont.columns)
```

### Step 13: Isolating the Target Variable

The `label` column is extracted into a new DataFrame (`df_cat`), which contains the target variable. This step separates the dependent variable from the feature variables for model training.


```python
df_cat=df[['label']]
```

### Step 14: Encoding the Target Variable

The `LabelEncoder` is used to convert the categorical labels in the `label` column into numerical values. This is necessary for classification algorithms like SVM, which require numerical input. The encoded labels are then updated in the `df_cat` DataFrame.


```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df_cat['label']=LE.fit_transform(df_cat['label'])
```

### Step 15: Combining Features and Target

The standardized features (`SS_X`) and the encoded target variable (`df_cat`) are concatenated into a single DataFrame (`df_new`). This combined dataset is ready for use in training the machine learning model.


```python
df_new=pd.concat([SS_X,df_cat],axis=1)
```

### Step 16: Checking for Missing Values in the Combined Dataset

The `isnull().sum()` method is used to check for any missing values in the combined dataset (`df_new`). This ensures that the data is clean and ready for model training.


```python
df_new.isnull().sum()
```

### Step 17: Dropping Missing Values

The `dropna()` method is used to remove any rows with missing values from the combined dataset (`df_new`). This step ensures that only complete data is used for model training.


```python
df_new.dropna(inplace=True)
```

## Step 18: Splitting Features and Target

The feature variables (`X`) are separated from the target variable (`Y`).

- **`X`**: Contains all the columns except for the `label`, representing the independent variables.
- **`Y`**: Contains the `label` column, which is the dependent variable (target) for classification.

This step prepares the data for model training.


```python
X=df_new.drop(df[['label']],axis=1)
Y=df_new['label']
```

### Step 19: Model Training and Evaluation

In this step, we perform model training and evaluation using cross-validation:

- **Data Splitting**: The dataset is split into training and testing sets (70% training, 30% testing) using `train_test_split()`.
- **SVM Model**: A Support Vector Machine (SVM) model with a polynomial kernel (`kernel='poly'`) is trained on the training data (`X_train`, `Y_train`).
- **Prediction**: The model makes predictions on both the training and testing sets.
- **Accuracy Calculation**: The accuracy score for both training and test sets is calculated using `accuracy_score()` and stored in separate lists for cross-validation.

Finally, the average training and test accuracy across 100 iterations is printed, providing an overall performance estimate of the model.


```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
training_acc=[]
test_acc=[]
for i in range(1,101):
  X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=i)
  model=SVC(kernel='poly')
  model.fit(X_train,Y_train)
  pred_Y_train=model.predict(X_train)
  pred_Y_test=model.predict(X_test)
  training_acc.append(accuracy_score(Y_train,pred_Y_train))
  test_acc.append(accuracy_score(Y_test,pred_Y_test))
print("cross validation training accuracy score",np.round(np.mean(training_acc),2))
print("cross validation test accuracy score",np.round(np.mean(test_acc),2))
```

### Step 20: Plotting Decision Regions

In this step, we plot the decision boundaries of the Support Vector Machine (SVM) model to visually understand how the model separates different classes based on the selected features:

- **Feature Selection**: We select two features (`meanfreq` and `sd`) for visualization. This is important because decision boundaries can only be plotted for 2D feature sets.
- **Model Training**: The SVM model with a linear kernel (`kernel='linear'`) is trained on the selected features (`X_subset`) and the target variable (`y_subset`).
- **Decision Boundary Plot**: The `plot_decision_regions()` function is used to visualize the decision boundaries. The target variable is converted to integers to satisfy the function's requirements.
- **Visualization**: The plot is customized with axis labels (`meanfreq` and `sd`) and a title to clearly indicate the decision regions.

This plot helps in understanding how the SVM model classifies the data based on the chosen features.


```python


from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

X_subset = X[['meanfreq', 'sd']]
y_subset = Y

model = SVC(kernel='linear')
model.fit(X_subset, y_subset)

y_subset = y_subset.astype(np.int_)  # Convert y to integers


plot_decision_regions(X_subset.values, y_subset.values, clf=model, legend=2)
plt.xlabel('meanfreq')
plt.ylabel('sd')
plt.title('Decision Regions')
plt.show()
```
