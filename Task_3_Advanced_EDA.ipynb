# 1.Maximum Product of Two Elements in an Array

```python
#1.Maximum Product of Two Elements in an Array
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        nums.sort()
        return (nums[-1]-1)*(nums[-2]-1)
```

```python
obj=Solution()
nums =[3,4,5,2]
print(obj.maxProduct(nums))
```

# 2.Count Number of Teams

```python
#2.Count Number of Teams
class Solution:
    def numTeams(self, rating: List[int]) -> int:
        n = len(rating)
        count = 0

        for j in range(n):
            leftLess = leftGreater = rightLess = rightGreater = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    leftLess += 1
                elif rating[i] > rating[j]:
                    leftGreater += 1

            for k in range(j + 1, n):
                if rating[k] < rating[j]:
                    rightLess += 1
                elif rating[k] > rating[j]:
                    rightGreater += 1

            count += leftLess * rightGreater + leftGreater * rightLess

        return count
```

```python
ob=Solution()
rating =[2,5,3,4,1]
print(ob.numTeams(rating))
```

# 3.Number of Students Doing Homework at a Given Time

```python
#3.Number of Students Doing Homework at a Given Time
class Solution:
    def busyStudent(self, startTime: List[int], endTime: List[int], queryTime: int) -> int:
        doingHomework = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                doingHomework += 1
        return doingHomework
```

```python
out=Solution()
startTime = [1,2,3]
endTime = [3,2,7]
queryTime = 4
print(out.busyStudent(startTime,endTime,queryTime))
```

# 4.Number of Steps to Reduce a Number to Zero

```python
#4.Number of Steps to Reduce a Number to Zero
class Solution:
    def numberOfSteps(self, num: int) -> int:
        count  = 0
        while num > 0:
            if num % 2 == 0:
                num //= 2
                count += 1
            else:
                num -= 1
                count += 1
        return count
```

```python
res=Solution()
num =14
print(res.numberOfSteps(num))
```

# 5.Counting Bits

```python
#5.Counting Bits
class Solution:
    def countBits(self, n: int) -> List[int]:
        a = []
        for i in range(0,n+1):
            a.append(bin(i).count("1"))
        return a
```

```python
ans=Solution()
n =2
print(ans.countBits(n))
```
