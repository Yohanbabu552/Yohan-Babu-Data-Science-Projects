# 2.Python-If-Else

```python
#2.Python-If-Else
n=int(input())
if (n % 2 != 0):
    print("Weird")
else:
    if (n>=2 and n<=5):
        print("Not Weird")
    elif (n>6 and n <= 20):
        print("Weird")
    else:
        print("Not Weird")
```

# 6.Write a Function

```python
#6.Write a Function
def is_leap(year):
    if (year%4==0 and year%100!=0)or (year%400==0):
        return True
    else:
        return False
year = int(input())
print(is_leap(year))
```
